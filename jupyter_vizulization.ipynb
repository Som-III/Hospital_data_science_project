```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import scipy.stats as stats
```

```python
df=pd.read_excel('insurance (1).xlsx')
```

```python
plt.figure(figsize=(5,4))
sns.histplot(data=df, x="HBA1C", bins=10, kde=True)
plt.title("Distribution of HBA1C")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.histplot(data=df, x="BMI", bins=10, kde=True)
plt.title("Distribution of BMI")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.boxplot(data=df, y="HBA1C")
plt.title("Box Plot of HBA1C")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 4))

sns.boxplot(data=df, y="BMI")
plt.title("Box Plot of BMI")
plt.tight_layout()
plt.show()
```

```python
print(df[["HBA1C", "BMI"]].describe())
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),cmap='rainbow',linewidth=1,annot=True,vmin=-1,vmax=1)
```
