```python
month_map={
    'OCT':10,
    'JUN':6,
    'AUG':8, 
    'SEP':9, 
    'DEC':12,
    'JUL':7, 
    'NOV':11
}
```

```python
final['Month']=final['Month'].map(month_map)
```

```python
final['Date']=pd.to_datetime(final[['Year','Month','Date']])
```

```python
def dating(row):
    return str(row['Date'])+"/"+str(row['Month'])+"/"+str(row['Year'])
```

```python
final['DATE'] = ''
final['DATE'] = final.apply(lambda x : dating(x), axis=1)
```
